<h3 style = "color:#04B404">3.4 Rangos </h3>

```python
dfAlcohol.head(10)
```

```python
dfAlcohol.dtypes
```

```python
dfAlcohol=dfAlcohol.drop(dfAlcohol.index[[1,0]])
```

```python
dfAlcohol.head(5)
```

```python
dfAlcohol.dtypes
```

```python
bins = np.linspace(min(dfAlcohol ["Consumo"]),max(dfAlcohol["Consumo"]),4)
bins
```

```python
group_names = ["bajo","medio","alto"]
dfAlcohol ["Consumo-rangos"] = pd.cut(dfAlcohol["Consumo"], bins, labels=group_names,include_lowest=True)
dfAlcohol[['Consumo','Consumo-rangos']].head(10)
```

```python
dfAlcohol ["Consumo-rangos"].value_counts()
```

```python
dfAlcohol.head(20)
```

<h3 style = "color:#ff0001"> 4 Visualización de Datos </h3>

```python
# agrupar los datos para luego graficar.
gfl = dfAlcohol.groupby (["Consumo-rangos"]).count()
gfl
```

<h3 style = "color:#001a33"> 1. ¿Como se distribuye los paises segun el consumo de Alcohol por rango de Consumo? </h3>

```python
gfl = gfl.iloc[:,:1]
gfl
```

```python
eje_x = ('bajo', 'medio', 'alto')
eje_y = gfl['Pais'].tolist()

## Crear la grafica

plt.bar(eje_x, eje_y,color='#4d94ff', alpha = 0.7)

#Leyenda en el eje y
plt.ylabel('Cantidad de Paises')

#Leyenda en el eje x
plt.xlabel('Rango de Consumo')

#Titulo al Grafico
plt.title('Consumo de Alcohol por paises')

#Cambia el tamaño del grafico
plt.figure(figsize=(10,30))

plt.show()
```

<h3 style = "color:#001a33"> 2. ¿Como se distribuyen la muestra de paises, segun los Continentes? </h3>

```python
gf2 = dfAlcohol.groupby(["Continet"]).count()
gf2 = gf2.iloc[:,:1]
gf2
```

```python
eje_x = ('Africa', 'Antarctica', 'Asia','Europe','North America','Oceania','South America')

eje_y = gf2['Pais'].tolist()

## Crear la grafica

plt.bar(eje_x, eje_y,color='#004d99', alpha = 1)

#Leyenda en el eje y
plt.ylabel('Cantidad de Paises')

#Leyenda en el eje x
plt.xlabel('Distribucion por Continentes')

#Titulo al Grafico
plt.title('Consumo de Alcohol por Continente')

#Cambia el tamaño del grafico
plt.figure(figsize=(10,30))

plt.show()
```

```python
<h3 style = "color:#001a33"> 3. ¿Distibución porcentual de la muestra de paises, segun los Continentes? </h3>
```

```python
gf3 = dfAlcohol.groupby(["Continet"]).count()
gf3 = gf3.iloc[:,:1]
gf3
```

```python
labels = 'Africa','Antarctica','Asia','Europe','North America','Oceania','South America'
sizes = gf3['Pais'].tolist()
explode = (0, 0.1,0,0,0,0,0)  

fig1, ax1 = plt.subplots()

ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',shadow=True, startangle=90)

ax1.axis('equal')

plt.show()
```

<h3 style = "color:#b36b00"> Eliminación de Valores Perdidos </h3>

```python
dfAlcohol.dropna(subset = ['Year'],axis = 0,inplace = True)
dfAlcohol.reset_index(drop=True,inplace=True)

dfAlcohol.dropna(subset = ['Total alcohol consumption per capita (liters of pure alcohol projected estimates 15+ years of age)'],axis = 0,inplace = True)
dfAlcohol.reset_index(drop=True,inplace=True)

dfAlcohol.dropna(subset = ['GDP per capita PPP (constant 2017 international $)'],axis = 0,inplace = True)
dfAlcohol.reset_index(drop=True,inplace=True)

dfAlcohol.dropna(subset = ['Population (historical estimates)'],axis = 0,inplace = True)
dfAlcohol.reset_index(drop=True,inplace=True)

dfAlcohol.dropna(subset = ['Consumo'],axis = 0,inplace = True)
dfAlcohol.reset_index(drop=True,inplace=True)

dfAlcohol.dropna(subset = ['Consumo-rangos'],axis = 0,inplace = True)
dfAlcohol.reset_index(drop=True,inplace=True)
```

```python
valoresPerdidos = dfAlcohol.isnull()
valoresPerdidos.head(5)

for column in valoresPerdidos.columns.values.tolist():
    print(column)
    print(valoresPerdidos[column].value_counts())
    print ("")
```
